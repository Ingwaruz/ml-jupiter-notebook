```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('bar_pass_prediction.csv')
```

```python
#1. Открыть набор данных и вывести техническую информацию о наборе данных .info()
df.info
```

```python
#2. Выбрать в качестве переменной - непрерывное значение
x = df["ugpa"]
# 2.1 Построить диаграмму плотности распределения (гистограмму). Как вы думаете, на какой вид распределения похож график?
plt.hist (x)
plt.grid()
plt.minorticks_on()
plt.show()
```

## **Вид распределение - нормальное**
Из полученной гистограммы можно сделать вывод, что у большинства студентов балакавриата средний балл средний балл 3.18-3.30 и лишь у небольшого числа студентов средний балл меньше 2.

```python
#2.2 Построить boxplot
x = df["ugpa"]
plt.boxplot(x, showmeans=True)
plt.minorticks_on()
plt.show()
```

Из полученного boxplot можно сделать вывод, что медиана - 3.2 балла

```python
#3. Выбрать несколько непрерывных переменных и провести визуализацию отношений между переменными
x = df["ugpa"]
y = df["lsat"]
#3.1 Построить диаграмму рассеивания, гистограмму (относительно двух переменных), матрицу корреляций

#диаграмма
plt.scatter(x, y, s=2)
```

```python
# Гистограмма
width = 0.35  
x = np.arange(len(df[:50]))  
fig, ax = plt.subplots(figsize=(40,10))
rects1 = ax.bar(x - width/2, df['ugpa'][:50], width, label='ugpa')
rects2 = ax.bar(x + width/2, df['lsat'][:50], width, label='lsat')

ax.set_xticks(x)
ax.legend()
```

```python
#матрица корреляций
df1 = df[['ugpa','lsat']].copy()
corr = df1.corr()
corr.style.background_gradient(cmap='coolwarm')
```

Вывод: чем выше оценка вступительного экзамена, тем выше средний балл студента


```python
#4. Выбрать несколько переменных, X - категориальное, Y - непрерывное
#4.1 Построить гистограмму Y относительно переменной X
sns.pairplot( df[["ugpa", "sex"]] )
```

```python
#4.2 Построить диаграмму размаха (boxplot) относительно переменной X
sns.boxplot(x='sex', y='ugpa', data=df, whis=10)

plt.xlabel('sex')
plt.ylabel('ugpa')
```
