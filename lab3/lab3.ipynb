```python
import pandas as pd
import numpy as np

# функция для вычисления евклидовой метрики между двумя векторами
def euclidean_distance(x, y):
    return np.sqrt(np.sum((x - y) ** 2))

temp_df = df.select_dtypes(include=['float64'])


# замена пропущенных значений с помощью евклидовой метрики
for i, row in temp_df.iterrows():
    for j in range(len(row)):
        if pd.isna(row[j]):
            distances = []
            for k, other_row in temp_df.iterrows():
                if k != i and not pd.isna(other_row[j]):
                    distances.append((k, euclidean_distance(row.fillna(0), other_row.fillna(0))))
            if distances:
                nearest = min(distances, key=lambda x: x[1])[0]
                temp_df.iat[i, j] = temp_df.iat[nearest, j]

print(temp_df)
```

Метод восстановление с помощью метрик наиболее подходящий, потому что получаем приближенное разные приближенные значения для пропущенных элементов, в других же способах берутся одинаковые средние/медиана значения.

```python
#2. Применить методы замены пропусков для номинальных: на основе моды (можно использовать более сложные методы, например предсказывания на основе других признаков, но если существует связь с этим(-и) признаками, например на основе тестов на значиомсть)
temp_df = df.select_dtypes(include=['category'])

for column in temp_df.columns.values.tolist():
    if pd.isnull(temp_df[column]).sum()>0:
        temp_df[column] = temp_df[column].fillna(temp_df[column].to_numpy()[0])
        print(temp_df[column])
        print('=============')
```

## **5. Применить поиск выбросов в числовых признаках (Не забываем про нормирование):**

```python
#(необходимо исполнить код из дополнения значений метрики)
#1. Провести поиск аномалий числовых переменных методами одного признака.
    #1.1. На основе простейшего метода из алгоритма 2 (с учетом нормальности)
temp_df2 = temp_df['highway_mileage']
# нормируем данные
normalized_data = (temp_df2 - temp_df2.mean()) / temp_df2.std()

# определяем границы диапазона, в котором данные считаются нормальными
lower_bound = normalized_data.mean() - (3 * normalized_data.std())
upper_bound = normalized_data.mean() + (3 * normalized_data.std())

# выводим значения, которые находятся за границами диапазона
anomalies = temp_df2.loc[(normalized_data < lower_bound) | (normalized_data > upper_bound)]
print(anomalies)
```

```python
#1.2. На основе методов без использование среднего и СКО
temp_df2 = temp_df['highway_mileage']
# вычисляем квартили
q1 = temp_df2.quantile(0.25)
q3 = temp_df2.quantile(0.75)

# вычисляем интерквартильный размах
iqr = q3 - q1

# определяем границы диапазона, в котором данные считаются нормальными
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# выводим значения, которые находятся за границами диапазона
anomalies = temp_df2.loc[(temp_df2 < lower_bound) | (temp_df2 > upper_bound)]
print(anomalies)
```

Метод без использование среднего и СКО подходит больше так как выводит также верхнее значения, одинственное в датафрейме, что можно считать отклонением

```python
#1.4. На основе критерия Шовене
from scipy.stats import shapiro

# создаем DataFrame с данными
temp_df2 = temp_df['highway_mileage']

# нормируем данные
normalized_df = (temp_df2 - temp_df2.mean()) / temp_df2.std()

# применяем критерий Шовене
stat, p = shapiro(normalized_df)

# устанавливаем уровень значимости
alpha = 0.05

# выводим результаты
if p > alpha:
    print('Данные распределены нормально')
else:
    print('Данные не распределены нормально')
    # находим аномальные значения
    anomalies = normalized_df[(normalized_df < -2) | (normalized_df > 2)]
    print('Аномальные значения:')
    print(anomalies)
```

```python
#3. Выбрать в качестве X и Y номинальные переменные

from scipy.stats import chi2_contingency
data = pd.read_csv('used_cars.csv')

# Подсчет частот каждой категории
observed_freq = pd.crosstab(data['brand'], data['engine_type'])

# Применение критерия хи-квадрат
chi2, p_val, dof, expected_freq = chi2_contingency(observed_freq)

print("Chi-square statistic: ", chi2)
print("P-value: ", p_val)
print("Degrees of freedom: ", dof)
print("Expected frequencies:\n", expected_freq)
```

Вывод будет содержать значения статистики хи-квадрат, p-value, степеней свободы и ожидаемые частоты. Если p-value меньше выбранного уровня значимости, то можно отвергнуть нулевую гипотезу и сделать вывод о наличии статистически значимой связи между номинальными переменными.
